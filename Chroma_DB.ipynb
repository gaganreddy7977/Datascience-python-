```python
documents = [
    "Azure Data Factory is used for building ETL pipelines",
    "Apache Spark is used for big data processing",
    "ChromaDB is a vector database for embeddings",
    "Sentence Transformers generate embeddings using transformers",
    "Azure Databricks is built on top of Apache Spark"
]

ids = [f"doc_{i}" for i in range(len(documents))]
```

# Using Embedding


```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
embeddings = model.encode(documents)

print("embedding shape :", embeddings.shape)
```

```python
import chromadb

client = chromadb.Client()
collection = client.create_collection(name="my_collection")

collection.add(
    documents=documents,
    embeddings=embeddings.tolist(),
    ids = ids
)

print("Documents strored in Chromadb")

query = "What is the use of ETL pipelines?"

query_embedding = model.encode([query])

results = collection.query(
    query_embeddings=query_embedding.tolist(),
    n_results=2
)


print("Query:",query)
print("Top matches:")
for doc in results["documents"][0]:
    print("-",doc)
```
